# **一、分类树**

```python
from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
```

```python
wine = load_wine()
```

```python
wine.data
```

```python
wine.data.shape
```

```python
wine.target
```

```python
import pandas as pd
pd.concat([pd.DataFrame(wine.data), pd.DataFrame(wine.target)], axis=1)
```

```python
wine.feature_names
```

```python
wine.target_names
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(wine.data, wine.target, test_size=0.3)
```

```python
Xtrain.shape
```

```python
Xtest.shape
```

```python
Ytrain
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy')
clf.fit(Xtrain, Ytrain)
score = clf.score(Xtest, Ytest)

score
```

```python
feature_name = ["酒精", "苹果酸", "灰", "灰的碱性", "镁", "总酚", "类黄酮", "非黄烧类酚类", "花青素", "颜色强度", "色调", "od280/od315稀释葡萄酒", "脯氨酸"]

import graphviz
dot_data = tree.export_graphviz(clf,
                                feature_names=feature_name,
                                class_names=["琴酒", "雪莉", "贝尔摩德"],
                                filled=True,
                                rounded=True
                                )
graph = graphviz.Source(dot_data)
graph
```

```python
clf.feature_importances_
```

```python
[*zip(feature_name, clf.feature_importances_)]
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy', random_state=22)
clf.fit(Xtrain, Ytrain)
score = clf.score(Xtest, Ytest)
score
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy',
                                  random_state=26,
                                  splitter='random')
clf.fit(Xtrain, Ytrain)
score = clf.score(Xtest, Ytest)
score
```

```python
import graphviz
dot_data = tree.export_graphviz(clf,
                                feature_names=feature_name,
                                class_names=["琴酒", "雪莉", "贝尔摩德"],
                                filled=True,
                                rounded=True
                                )
graph = graphviz.Source(dot_data)
graph
```

```python
score_train = clf.score(Xtrain, Ytrain)
score_train
```

**剪枝参数调优**

```python
clf = tree.DecisionTreeClassifier(criterion='entropy',
                                  random_state=26,
                                  splitter='random',
                                  max_depth=5,
                                  #min_samples_leaf=10,
                                  #min_samples_split=10
                                  )
clf.fit(Xtrain, Ytrain)
dot_data = tree.export_graphviz(clf,
                                feature_names=feature_name,
                                class_names=["琴酒", "雪莉", "贝尔摩德"],
                                filled=True,
                                rounded=True
                                )
graph = graphviz.Source(dot_data)
graph
```

```python
score = clf.score(Xtest, Ytest)
score
```

```python
import matplotlib.pyplot as plt
test = []
for i in range(10):
    clf = tree.DecisionTreeClassifier(criterion='entropy',
                                      max_depth=i+1,
                                      random_state=26,
                                      splitter='random'
                                      )
    clf.fit(Xtrain, Ytrain)
    score = clf.score(Xtest, Ytest)
    test.append(score)

plt.plot(range(1, 11), test, color='r', label="max_depth")
plt.legend()
plt.show()
```
